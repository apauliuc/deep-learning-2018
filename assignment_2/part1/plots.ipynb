```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(13, 8))
plt.plot(x, rnn_accuracy, 'b', label='RNN')
plt.xticks(x, sizes)
plt.xlabel("Sequence length")
plt.ylabel("Accuracy")
# plt.title("RNN accuracy")
plt.savefig('rnn.png')
# plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(13, 8))
plt.plot(x, lstm_accuracy, 'r', label='LSTM')
plt.xticks(x, sizes)
plt.xlabel("Sequence length")
plt.ylabel("Accuracy")
# plt.title("LSTM accuracy")
# plt.savefig('lstm.png')
# plt.legend()
plt.show()
```
